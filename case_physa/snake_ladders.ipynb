```python
import pandas as pd
import numpy as np
from random import randint
```

```python
snakes = pd.DataFrame(
    [[12,2],
     [14,11],
     [17,4],
     [31,19],
     [35,22]],
     columns = ['head', 'tail']
)
snakes
```

```python
ladders = pd.DataFrame(
    [[3,16],
     [5,7],
     [15,25],
     [18,20],
     [21,32]],
     columns = ['bottom', 'top']
)
ladders
```

```python
position = [1,1]
win = [0,0]

turn_count = 0
player = 0
while True:
    turn_count += 1
    dice = randint(1, 6)
    print(f'Jogador {player} tira o valor {dice}')

    position[player] += dice
    print(f'Posições atualizadas: {position}')

    if position[player] in snakes['head'].values:
        print(f'Jogador {player} cai da casa {position[player]} para a casa {}')
        position[player] = 
        snakes[snakes['head'] == position[player]].index


    if position[player] >= 36:
        win[player] += 1
        print(f'O jogador {player} ganhou!')
        print(f'Número de rodadas: {turn_count}')
        break
    
    if player == 0:
        player = 1
    else:
        player = 0
    

```
